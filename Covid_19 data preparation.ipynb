```python
import pandas as pd
```

```python
covid_19_confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
covid_19_death_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
covid_19_recovered_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
covid_19_confirmed_df.head()
```

```python
covid_19_confirmed_df = covid_19_confirmed_df.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
```

```python
covid_19_confirmed_df.head()
```

```python
covid_19_confirmed_df = covid_19_confirmed_df.rename({'variable':'Date','value':'Confirmed'},axis = 'columns')
```

```python
covid_19_confirmed_df.head()
```

```python
covid_19_death_df.head()
```

```python
covid_19_death_df = covid_19_death_df.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
```

```python
covid_19_death_df = covid_19_death_df.rename({'variable':'Date','value':'Death'},axis = 'columns')
```

```python
covid_19_death_df.head()
```

```python
covid_19_recovered_df.head()
```

```python
covid_19_recovered_df = covid_19_recovered_df.melt(id_vars=['Province/State','Country/Region','Lat','Long'])
```

```python
covid_19_recovered_df.head()
```

```python
covid_19_recovered_df.isnull().any()
```

```python
covid_19_recovered_df = covid_19_recovered_df.rename({'variable':'Date','value':'Recovered'},axis = 'columns')
```

```python
covid_19_recovered_df.head()
```

```python
covid_19_recovered_df.isnull().any()
```

```python
com_df = [covid_19_confirmed_df,covid_19_death_df,covid_19_recovered_df]
com_df = [df.set_index(['Province/State','Country/Region','Lat','Long','Date']) for df in com_df]
com_df = com_df[0].join(com_df[1:])
com_df.head()
```

```python
com_df.isnull().sum()
```

```python
com_df = com_df.reset_index()
```

```python
com_df.head()
```

```python
com_df.isnull().any()
```

```python
#Change data type to numeric for numbers
com_df[['Lat','Long','Confirmed','Death','Recovered']] = com_df[['Lat','Long','Confirmed','Death','Recovered']].apply(pd.to_numeric)
```

```python
com_df[['Date']] = com_df[['Date']].apply(pd.to_datetime)
```

```python
com_df.dtypes
```

```python
com_df.head()
```

```python
com_df = com_df.rename({'Province/State':'State','Country/Region':'Country'},axis = 'columns')
```

```python
com_df.head()
```

```python
com_df.isnull().any()
```

```python
com_df.count()
```

```python
com_df.set_index('Country').filter(like='India',axis=0)
```

```python
com_df.to_csv('covid_19_dataset.csv')
```
